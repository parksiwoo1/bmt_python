```python
fp=open('지역평균기온.txt')
data=fp.readlines()
fp.close()

dt={}
for line in data:
  line=line.replace('\n','')
  items=line.split(',')
  dt[items[0]]=int(items[1])

area=list(dt.keys())
temp=list(dt.values())

import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.plot(area,temp,'g--')
plt.scatter(area,temp,color='green')
plt.xticks(rotation=30)
plt.title('지역평균기온')
plt.show()
```
