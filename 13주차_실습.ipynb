```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
html=requests.get('https://www.bbq.co.kr/menu/menuList.asp')
soup=bs(html.text,'html.parser')
```

```python
name=soup.select('div.info > p.name')
```

```python
name
names=[]
for i in name:
  names.append(i.text)
```

```python
names
```

```python
price=soup.select('div.info > p.pay')
prices=[]
for i in price:
  prices.append(i.text)
```

```python
prices
```

```python
header = ['메뉴', '가격']
df = pd.DataFrame([x for x in zip(names,prices)], columns=header)
print(df)

# 문자열인 가격을 숫자로 바꿔주는 작업. 평균을 구하려면 데이터타입이 INT여야한다.
df['가격'] = df['가격'].apply(lambda x: x.replace('원', ''))
df['가격'] = df['가격'].apply(lambda x: x.replace(',', ''))
df = df.astype({'가격': 'int'})
avg = df['가격'].mean()
std = df['가격'].std()

print('전체 평균가  : %d 원'%avg)
print('표준 편차   : ', std)
```
